# Reproducing Google's Surface-Code Scaling (Nature 2023)

We extract the logical error per cycle (epsilon) for distances 3, 5 and 7 by fitting the decay of
logical fidelity with the number of rounds, then compute the suppression factor Lambda. Below
threshold we reproduce the paper's central claim: larger distance, smaller epsilon.

```python
%matplotlib inline
import matplotlib.pyplot as plt

from grepro.experiments import lambda_factor
from grepro.figures import plot_epsilon_vs_distance, plot_fidelity_decay
from grepro.repro import run_reproduction

results = run_reproduction(p=0.004, shots=40000, seed=2023)
by_d = {r.distance: r.epsilon for r in results}
for d, eps in sorted(by_d.items()):
    print(f"d={d}: epsilon={eps:.4f}")
print(f"Lambda(3->5) = {lambda_factor(by_d[3], by_d[5]):.2f}")
print(f"Lambda(5->7) = {lambda_factor(by_d[5], by_d[7]):.2f}")
```

```python
plot_fidelity_decay(results)
plt.tight_layout()
plt.show()

plot_epsilon_vs_distance(results)
plt.tight_layout()
plt.show()
```
